```python
import pandas as pd
data = {
    'Order_ID':['01','02','03','04','05','06','07','08','09','10'],

    'Category':['Electronics','Clothing','Electronics','Grocery','Clothing',
                'Electronics','Grocery','Clothing','Electronics','Grocery'],

    'Payment_Method':['UPI','Card','UPI','Cash','Card',
                       'UPI','Cash','Card','UPI','Cash'],

    'Delivery_Type':['Express','Standard','Express','Standard','Express',
                      'Standard','Express','Standard','Express','Standard'],

    'Customer_Type':['New',"Returning","Returning","New","Returning",
                      'New','Returning','New','Returning','New']
}

df = pd.DataFrame(data)
df
```

```python
#1 Frequency Count
#Count how many orders in each category
print(df['Category'].value_counts())
print()
print(df['Payment_Method'].value_counts())
print()
print(df['Customer_Type'].value_counts())
print()
print(df['Delivery_Type'].value_counts())
```

```python
#2 Unique Values
#Display all unique categories
print(df['Category'].unique())
print()
print(df['Payment_Method'].unique())
print()
print(df['Customer_Type'].unique())
print()
print(df['Delivery_Type'].unique())
```

```python
#3 Number of unique values
print("Category :",df['Category'].nunique())
print()
print("Payment Method ;",df['Payment_Method'].nunique())
print()
print("Customer Type :",df['Customer_Type'].nunique())
print()
print("Delivery Type:",df['Delivery_Type'].nunique())

```

```python
#4 Cross Tabulation
# Category VS Payment Method
print(pd.crosstab(df['Category'],df['Payment_Method']))
```

```python
#5 Cross Tabulation
# Customer Type VS Delivery Method
print(pd.crosstab(df['Delivery_Type'],df['Customer_Type']))
```

```python
#6 Percentage Distribution
print(df['Category'].value_counts(normalize=True)*100)
```

```python
#7 Filtering Data
print(df[df['Category']=='Electronics'])
```

```python

#8 Grouping
#Order by category and payment method
print(df.groupby('Category')['Payment_Method'].value_counts())
```

```python

#for me
# grouping
#order by category and payment method
print(df.groupby(['Category','Payment_Method']).value_counts())
```

```python
#9 Sorting Data
sorted_df = df.sort_values(by='Order_ID')
sorted_df
```

```python
# Import NumPy library
import numpy as np
```

```python
# Creating arrays
a = np.array([10, 20, 30, 40])
b = np.array([[1, 2, 3],
              [4, 5, 6]])

print("Array A:", a)
print("Array B:\n", b)
```

```python
# Checking dimensions of arrays

print("Dimension of Array A:", a.ndim)
print("Dimension of Array B:", b.ndim)

```

```python
# Display shape (rows, columns)

print("Shape of Array A:", a.shape)
print("Shape of Array B:", b.shape)

```

```python
# Checking data types of elements

print("Data type of elements in A:", a.dtype)
print("Data type of elements in B:", b.dtype)
```

```python
# Creating a zero matrix

zero_matrix = np.zeros((3, 4))
print(zero_matrix)
```

```python

# Creating a matrix filled with ones

ones_matrix = np.ones((2, 3))
print(ones_matrix)

```

```python

# Creating an identity matrix

identity_matrix = np.eye(3)
print(identity_matrix)

```

```python
# Creating array using arange()

range_array = np.arange(1, 11, 2)
print(range_array)
```

```python
# Performing arithmetic operations

c = a * 2
d = a + 5

print("Array A multiplied by 2:", c)
print("Array A plus 5:", d)
```

```python
# Calculating mean

print("Mean of Array A:", np.mean(a))
print("Mean of Array B:", np.mean(b))
```

```python
# Calculating median

print("Median of Array A:", np.median(a))
print("Median of Array B:", np.median(b))
```

```python
# Aggregate operations on Array A

print("Operations on Array A")
print("Maximum:", np.max(a))
print("Minimum:", np.min(a))
print("Sum:", np.sum(a))

# Aggregate operations on Array B

print("\nOperations on Array B")
print("Maximum:", np.max(b))
print("Minimum:", np.min(b))
print("Sum:", np.sum(b))

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = {

        'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday' ],
        'Study_Hours': [4, 6, 3, 5,3.5],
        'Marks': [50, 60, 55, 70, 65],
        'Attendance':[60, 65, 70, 75, 80],
        'Sleep_Hours': [7, 8, 6, 9, 7] ,
        'Assignment_Completed': [10, 15, 12, 18, 1]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Line chart

plt.plot(df['Day'],df['Study_Hours'],marker='o')
plt.show()
```

```python
#  Line chart

plt.plot(df['Day'],df['Study_Hours'],marker='o')
plt.title('Study Hours Over Days')
plt.xlabel('Day')
plt.ylabel('Study Hours')
```

```python
# Advanced Line Chart
plt.figure(figsize=(7,4))

plt.plot(df['Day'],df['Study_Hours'],marker='o',label='Study Hours')
plt.plot(df['Day'],df['Marks'],marker='o',label='Marks')

plt.title('Study Hours and Marks Trends')
plt.xlabel('Day')
plt.ylabel('Value')
plt.legend()
plt
```

```python
# Bar Chart

plt.bar(df['Day'],df['Study_Hours'],color='pink')
plt.title('Study Hours Over Days')
plt.xlabel('Day')
plt.ylabel('Study Hours')
plt.show()
```

```python
# Compare study hours and marks

# Bar Chart

plt.bar(df['Study_Hours'],df['Marks'],color='pink')
plt.title('Marks over study hours')
plt.xlabel('Study_Hours')
plt.ylabel('Marks')
plt.show()
```

```python
# Advanced Bar Graphs

bars = plt.bar(df['Day'],df['Study_Hours'],color='pink')
for bar in bars :
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,y,str(y),ha='center',va='bottom')
plt.title('Study Hours Over Days')
plt.xlabel('Day')
plt.ylabel('Study Hours')
plt.grid(axis='y')
plt.show()
```

```python
# Scatter Plot (Relationship)

import matplotlib.pyplot as plt
import pandas as pd

data = {
        'Day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday' ],
        'Study_Hours': [4, 6, 3, 5,3.5],
        'Marks': [50, 60, 55, 70, 65],
        'Attendance':[60, 65, 70, 75, 80],
        'Sleep_Hours': [7, 8, 6, 9, 7] ,
        'Assignment_Completed': [10, 15, 12, 18, 1]
}
df = pd.DataFrame(data)

plt.scatter(df['Study_Hours'],df['Marks'])

plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
# Conditional scatter plot

df['Result'] = ['Fall','Pass','Fall','Pass','Pass']
colors = ['red' if r=='Fall' else 'green' for r in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],c=colors)

plt.title('Study Hours vs Marks')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show
```

```python
# Line plot

sns.lineplot(x='Day',y='Study_Hours',data=df)
plt.title('Seaborn library plot')
plt.xlabel('Day')
plt.ylabel('Study Hours')
plt.show()
```

```python
# Advanced Line Chart
plt.figure(figsize=(7,4))

plt.plot(df['Day'], df['Study_Hours'],
         marker='o', color='green', label='Study Hours')

plt.plot(df['Day'], df['Marks'],
         marker='s', color='red', label='Marks')

plt.title("Study Hours and Marks Trend")
plt.xlabel("Day")
plt.ylabel("Values")
plt.legend()
plt.show()
```

```python
#2. Bar Chart (Comparison)

plt.bar(df['Day'], df['Marks'], color='green')

plt.title("Marks Comparison")
plt.xlabel("Day")
plt.ylabel("Marks")

plt.show()
```

```python
#Advanced Bar Graph
bars = plt.bar(df['Day'], df['Marks'], color='red')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y),
             ha='center', va='bottom')

plt.title("Marks Comparison")
plt.xlabel("Day")
plt.ylabel("Marks")
plt.grid(axis='y')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(len(df['Day']))
width = 0.35

plt.bar(x - width/2, df['Study_Hours'], width, label='Study Hours')
plt.bar(x + width/2, df['Marks'], width, label='Marks')

plt.xticks(x, df['Day'])

plt.title("Study Hours vs Marks")
plt.xlabel("Day")
plt.ylabel("Values")
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data (more values for better histogram)
marks = [50, 60, 55, 70, 65, 75, 80, 85, 90, 45, 68, 72, 77, 82, 88]

# Calculate mean
mean_value = np.mean(marks)

plt.figure(figsize=(7,4))

plt.hist(marks, bins=6, color='skyblue', edgecolor='black')

# Mean line
plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label='Mean')

plt.title("Advanced Histogram of Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.legend()

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
# Dataset-2
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North',
               'South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics',
                 'Food','Clothing','Electronics','Food']
}

df = pd.DataFrame(data)
print(df)
```

```python
#Line Plot
sns.lineplot(x='Day', y='Sales', data=df)
plt.title("Seaborn Line Plot")
plt.show()

```

```python
sns.barplot(x='Day', y='Profit', data=df)
plt.title("Seaborn Bar Plot")
plt.show()
```

```python
#Histogram
sns.histplot(df['Sales'], bins=5)
plt.title("Seaborn Histogram")
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("Seaborn Scatter Plot")
plt.show()
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize

text = "Natural language processing is interesting"

tokens = word_tokenize(text)

print(tokens)
```

```python
# Sentence Tokenization
from nltk.tokenize import sent_tokenize

text = "Python is easy. It is widely used in data science"
sentence = sent_tokenize(text)

print(sentence)
```

```python
# Stop word removal
# Stop words such as is , the,and,in are removed

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))

text = "Python is easy. It is widely used in data science"
words = word_tokenize(text)


filtered_words = [word for word in words if word.lower() not in stop_words]

print(filtered_words)
```

```python
# Stemming

# Stemming reduces words to their root form

from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

words = ["running", "runs", "ran","playing","watched","studies","coding"]

for w in words:
  stemmed_word = stemmer.stem(w)
  print(stemmed_word)
```

```python
# Lemmatization
# Lemmatization converts words to their dictionary form.

from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("running"))
print(lemmatizer.lemmatize("runs"))
print(lemmatizer.lemmatize("ran"))
print(lemmatizer.lemmatize("playing"))
print(lemmatizer.lemmatize("watched"))
print(lemmatizer.lemmatize("studies"))
print(lemmatizer.lemmatize("coding"))
```

```python
# Part-of-Speech POS
# POS tagging identifies grammitical roles of words.

import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
from nltk import pos_tag

words = word_tokenize("Python is a powerful programming language")

pos_tags = pos_tag(words)

print(pos_tags)

```

```python
# Word frequency count
from nltk.probability import FreqDist

words = word_tokenize("Python is a powerful programming language")

fd = FreqDist(words)

print(fd.most_common())
```
